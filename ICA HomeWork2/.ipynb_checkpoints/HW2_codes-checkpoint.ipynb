```python
solTrainXcontinous = solTrainXtrans.iloc[:,208:]
```

```python
corr = solTrainXcontinous.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(18, 12))

sns.heatmap(
    corr,
    mask=mask,
    annot=True,
    fmt=".2f",
    linewidths=.5,
    ax=ax, 
    vmin=-1, vmax=1, 
    cmap="viridis",
    cbar_kws={"shrink": .7})
```
